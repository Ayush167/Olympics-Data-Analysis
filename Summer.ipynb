# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
```

```python
df = pd.read_csv("summer.csv")
```

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
arr =[]
data = df[df['Medal'] == 'Gold']
for Sport in data['Sport'].unique():
    arr.append([Sport, len(data[data['Sport'] == Sport])])
arr = pd.DataFrame(arr, columns = ['Sport', 'Gold'])

arr=arr.sort_values(by = 'Gold', ascending = False).head()

```

```python
arr.plot(x='Sport', y='Gold', kind='bar', figsize = (5,5))
```
